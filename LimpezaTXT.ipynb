```python
# IMPORTANDO BIBLIOTECAS UTILIZADAS

import numpy as np
import pandas as pd
```

```python
# EXTRAÇÃO DE DADOS DE ARQUIVO TXT UTILIZANDO A BIBLIOTECA NUMPY

dados = np.loadtxt('database/alunos matriculados if.txt', delimiter=',', dtype=str, skiprows=1, encoding='utf-8')
dados
```

```python
# LIMPEZA DOS DADOS QUE VIERAM COM ASPAS ENTRE OS DELIMITADORES

for i in range(0, len(dados)):
  for j in range(0, len(dados[0])):
    dados[i][j] = dados[i][j].strip('\"')

dados
```

```python
# PRIMEIRO MÉTODO: EXTRAÇÃO DE DADOS REALIZADA POR ITERAÇÃO DE TUPLAS

unid = []
mod = []
tcurso = []
toferta = []
ano = []
status = []
quant = []


for i in range(0, len(dados)):
    unid.append(dados[i][0])
    mod.append(dados[i][1])
    tcurso.append(dados[i][2])
    toferta.append(dados[i][3])
    ano.append(dados[i][4])
    status.append(dados[i][5])
    quant.append(dados[i][6])
    
base = {
           'Unidade': unid,
           'Modalidade': mod,
           'Tipo_Curso': tcurso,
           'Tipo_Oferta': toferta,
           'Ano': ano,
           'Status': status,
           'Quantidade': quant
        }
```

```python
# FORMADAS AS TUPLAS DE CADA COLUNA, FORAM AGLUTINADAS EM UM DATAFRAME DO PANDAS

dataset1 = pd.DataFrame(base)
dataset1
```

```python
# SEGUNDO MÉTODO: COM O PANDAS O DATAFRAME PODE SER FEITO SEM A FORMAÇÃO DAS TUPLAS, APENAS NOMEANDO AS COLUNAS DO ARRAY NUMPY

dataset2 = pd.DataFrame(dados,
                       columns = ['Unidade', 'Modalidade', 'Tipo_Curso', 'Tipo_Oferta', 'Ano', 'Status', 'Quantidade'])
dataset2
```

```python
# O MESMO RESULTADO COM DOIS MÉTODOS DIFERENTES

dataset1.info()
print('\n')
dataset2.info()
```
